```python
import numpy as np
import pandas as pd
import xlrd
import matplotlib.pyplot as plt
```

```python
data_path = 'E:\\Documents\\Time Series\\Project\\data\\data_VNIdex_2021_2023.csv'
```

```python
data_raw = pd.read_csv(data_path)
```

```python
data_raw
```

```python
data_raw['Date'] = pd.to_datetime(data_raw['Date'], format='%Y/%m/%d')
```

```python
# data.to_csv('data_VNIdex_2021_2023.csv', index=False)
```

```python
data_raw
```

```python
data = data_raw[['Date', 'Close']]
```

```python
data
```

```python
# data.set_index('Date', inplace=True)
data = data.asfreq("D")
```

```python
data
```

```python
data= data.dropna()
```

```python
data
```

```python
data['Close'] = data['Close'].str.replace(',', '').astype(float)
```

```python


# Plot 'Close' column
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Close'], marker='o', linestyle='-')
plt.title('Close Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
